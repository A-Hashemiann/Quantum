```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
qA = QuantumRegister(1,name='alice')
qB = QuantumRegister(1,name='bob')
c = ClassicalRegister(2)
qc = QuantumCircuit(qA,qB,c)
qc.draw('mpl')
```

```python
qc.h(qA[0])
qc.cx(qA[0],qB[0])
qc.barrier()
qc.draw('mpl')
```

```python
message = '11'
if message == '00':
    qc.i(qA[0])
elif message == '01':
    qc.x(qA[0])
elif message == '10':
    qc.z(qA[0])
elif message == '11':
    qc.x(qA[0])
    qc.z(qA[0])
qc.barrier
qc.draw('mpl')
```

```python
qc.cx(qA[0],qB[0])
qc.h(qA[0])
qc.barrier()
qc.draw('mpl')
```

```python
qc.measure(qA[0],c[0])
qc.measure(qB[0],c[1])
qc.draw('mpl')
```
