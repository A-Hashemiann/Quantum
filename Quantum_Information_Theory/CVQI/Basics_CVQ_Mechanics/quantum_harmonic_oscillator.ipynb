```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
h=1
m=1
V=0.5
psi=0.01
dpsi=0
dx=0.01

E=0.5
#V*x**2 = E + 10V
a=np.sqrt(10+E/V)
x=-a


xlist=[]
psilist=[]

while x<=a:
    ddpsi=2*m/h**2*(V*x**2-E)*psi
    dpsi=dpsi+ddpsi*dx
    psi=psi+dpsi*dx
    x=x+dx
    xlist.append(x)
    psilist.append(psi)
    plt.plot(xlist,psilist)
            
        
```

This code numerically solves the Schrödinger equation for a quantum harmonic oscillator using a simple Euler integration method
## `ddpsi`
Computes the second derivative of the wavefunction using the Schrödinger equation:

$
\frac{d^2 \psi}{dx^2} = \frac{2m}{\hbar^2} (Vx^2 - E) \psi
$

Here $( V \cdot x^2 )$ is the potential energy of the harmonic oscillator.

## `dpsi`
Updates the first derivative of the wavefunction using Euler's method:

$
\frac{d\psi}{dx} \approx \frac{d\psi}{dx} + \frac{d^2 \psi}{dx^2} \cdot dx
$

## `psi`
Updates the wavefunction using Eulers method:
$
\psi \approx \psi + \frac{d\psi}{dx} \cdot dx
$

## `x`
Increments the position by \( dx \).

## `xlist` and `psilist`
Store the updated values of `x` and `psi`.
