```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
h=1
m=1
V=0.5
psi=0.01
dpsi=0
dx=0.01

E=0.5
#V*x**2 = E + 10V
a=np.sqrt(10+E/V)
x=-a


xlist=[]
psilist=[]

while x<=a:
    ddpsi=2*m/h**2*(V*x**2-E)*psi
    dpsi=dpsi+ddpsi*dx
    psi=psi+dpsi*dx
    x=x+dx
    xlist.append(x)
    psilist.append(psi)
    plt.plot(xlist,psilist)
            
        
```

This code numerically solves the Schrödinger equation for a quantum harmonic oscillator using a simple Euler integration method
## `ddpsi`
Computes the second derivative of the wavefunction using the Schrödinger equation:

$
\frac{d^2 \psi}{dx^2} = \frac{2m}{\hbar^2} (Vx^2 - E) \psi
$

Here $( V \cdot x^2 )$ is the potential energy of the harmonic oscillator.

## `dpsi`
Updates the first derivative of the wavefunction using Euler's method:

$
\frac{d\psi}{dx} \approx \frac{d\psi}{dx} + \frac{d^2 \psi}{dx^2} \cdot dx
$

## `psi`
Updates the wavefunction using Eulers method:
$
\psi \approx \psi + \frac{d\psi}{dx} \cdot dx
$

## `x`
Increments the position by \( dx \).

## `xlist` and `psilist`
Store the updated values of `x` and `psi`.

```python
E=0
psi=1
count=1
xvalues=[]
eigenfunctions=[]
eigenenergies=[]
nmax=10

while count<=nmax:
    psi=1
    while abs(psi)>0.02:
        psi=0.01
        dpsi=0.0
        E=E+0.05
        a=np.sqrt(10+E/V)
        x=-a
        xlist=[]
        psilist=[]

        while x<=a:
            ddpsi=2*m/h**2*(V*x**2-E)*psi
            dpsi=dpsi+ddpsi*dx
            psi=psi+dpsi*dx
            x=x+dx
            xlist.append(x)
            psilist.append(psi)
    
    eigenfunctions.append([psilist])
    eigenenergies.append(E)
    xvalues.append(xlist)
    count=count+1
    E=E*1.1
```

## Hermite Polynomials 


## Mathematical Definition  

$
H_0(x) = 1, \quad H_1(x) = 2x
$

$
H_n(x) = 2x H_{n-1}(x) - 2(n-1) H_{n-2}(x), \quad \text{for } n \geq 2
$

Alternatively, they can be expressed using Rodrigues' formula:

$
H_n(x) = (-1)^n e^{x^2} \frac{d^n}{dx^n} e^{-x^2}
$

The Hermite polynomials up to $( n = 4 )$:

$
H_0(x) = 1
$

$
H_1(x) = 2x
$

$
H_2(x) = 4x^2 - 2
$

$
H_3(x) = 8x^3 - 12x
$

$
H_4(x) = 16x^4 - 48x^2 + 12
$

---

1. **Orthogonality**:  
   $
   \int_{-\infty}^{\infty} e^{-x^2} H_m(x) H_n(x) \,dx = 0, \quad \text{for } m \neq n
   $
2. **Differential Equation**:  
   $
   H_n''(x) - 2x H_n'(x) + 2n H_n(x) = 0
   $
3. **Weight Function**: The Hermite polynomials satisfy the **weight function** $( w(x) = e^{-x^2} )$.

Computing the nth Hermite polynomial $H_n(x)$ using recursion

```python
import math
import matplotlib.pyplot as plt

def hermite_polynomial(n, x):
    if n == 0:
        return 1
    elif n == 1:
        return 2 * x
    else:
        return 2 * x * hermite_polynomial(n - 1, x) - 2 * (n - 1) * hermite_polynomial(n - 2, x)
```
