```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
h=1
m=1
V=0.5
psi=0.01
dpsi=0
dx=0.01

E=0.5
#V*x**2 = E + 10V
a=np.sqrt(10+E/V)
x=-a


xlist=[]
psilist=[]

while x<=a:
    ddpsi=2*m/h**2*(V*x**2-E)*psi
    dpsi=dpsi+ddpsi*dx
    psi=psi+dpsi*dx
    x=x+dx
    xlist.append(x)
    psilist.append(psi)
    plt.plot(xlist,psilist)
            
        
```
