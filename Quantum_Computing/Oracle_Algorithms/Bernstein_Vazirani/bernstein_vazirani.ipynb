```python
secret = '100101'
```

```python
n = len(secret)
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
q = QuantumRegister(n+1)
c = ClassicalRegister(n)
qc = QuantumCircuit(q,c)
```

```python
qc.x(q[n])
qc.barrier()
qc.draw('mpl')
```

```python
for i in range(n+1):
    qc.h(q[i])
qc.barrier()
qc.draw('mpl')
```

```python
for i in range(n):
    if secret[i]=='1':
        qc.cx(q[i],q[n])
qc.barrier()
qc.draw('mpl')
```
