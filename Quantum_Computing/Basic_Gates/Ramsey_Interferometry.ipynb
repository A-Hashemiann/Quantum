```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
q = QuantumRegister(1)
c = ClassicalRegister(1)
qc = QuantumCircuit(q,c)
```

```python
qc.draw('mpl')
```

```python
from math import pi
```

```python
qc.h(q[0])
qc.p(pi,q[0])
```

```python
qc.draw('mpl')
```

```python
from qiskit.quantum_info import Statevector
```

```python
psi = Statevector(qc)
psi.draw('latex')
```

```python
from qiskit.visualization import plot_state_qsphere
```

```python
plot_state_qsphere(qc)
```

```python
qc.h(q[0])
qc.draw('mpl')
```

```python
qc.measure(q[0],c[0])
qc.draw('mpl')
```
