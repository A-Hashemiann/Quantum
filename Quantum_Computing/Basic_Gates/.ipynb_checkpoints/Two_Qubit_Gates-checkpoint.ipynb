```python
psi3 = Statevector(qc)
psi3.draw('latex')
```

```python
qc.cx(q[0],q[1])
qc.draw('mpl')
```

```python
psi4 = Statevector(qc)
psi4.draw('latex')
```

```python
q_1= QuantumRegister(2)
qc0 = QuantumCircuit(q_1)
qc1 = QuantumCircuit(q_1)
qc2 = QuantumCircuit(q_1)
qc3 = QuantumCircuit(q_1)
```

```python
from qiskit.quantum_info import Statevector
```

```python
qc0.h(q_1[0])
qc0.cx(q_1[0],q_1[1])
qc0.draw('mpl')
```

```python
phi_plus = Statevector(qc0)
phi_plus.draw('latex')
```

```python
qc1.h(q_1[0])
qc1.cx(q_1[0],q_1[1])
qc1.z(q_1[0])
qc1.draw('mpl')
```

```python
phi_minus = Statevector(qc1)
phi_minus.draw('latex')
```

```python
qc2.h(q_1[0])
qc2.cx(q_1[0],q_1[1])
qc2.x(q_1[1])
qc2.draw('mpl')
```

```python
psi_plus = Statevector(qc2)
psi_plus.draw('latex')
```

```python
qc3.h(q_1[0])
qc3.cx(q_1[0],q_1[1])
qc3.z(q_1[0])
qc3.x(q_1[1])
qc3.draw('mpl')
```

```python
psi_minus = Statevector(qc3)
psi_minus.draw('latex')
```

```python
display(qc0.draw('mpl'),qc1.draw('mpl'),qc2.draw('mpl'),qc3.draw('mpl'))
```

```python
from qiskit.quantum_info import DensityMatrix
```

```python
rho_plus = DensityMatrix(qc0)
rho_plus.draw('latex',prefix = '\\rho_{AB}=')
```

```python
from qiskit.quantum_info import partial_trace
```

```python
rho_A = partial_trace(rho_plus,[0])
rho_B = partial_trace(rho_plus,[1])
```

```python
rho_A.draw('latex',prefix='\\rho_A=')
```

```python
rho_B.draw('latex',prefix='\\rho_B=')
```

```python
rho_A.purity()
```

```python
rho_B.purity()
```
